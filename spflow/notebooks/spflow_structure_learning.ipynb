## Structure learning using SPFlow on the LUCAS dataset

* Sum Product Flow: An Easy and Extensible Library for Sum-Product Networks
* https://github.com/SPFlow/SPFlow

```python
import spn
import numpy as np
from spn.algorithms.LearningWrappers import learn_mspn, learn_structure
from spn.io.Graphics import plot_spn
from spn.structure.Base import Context
from spn.structure.StatisticalTypes import MetaType
```

### Read in the dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("/home/queensgambit/Desktop/WiSe_18_19/PGM/project/PGM-Causal-Reasoning/data/LUCAS/lucas0_train.csv")
```

### Set the training set

```python
train_data = df.values
```

### Define the context of the MSPN

```python
ds_context = Context(meta_types=[MetaType.DISCRETE] * df.shape[1])
ds_context.add_domains(train_data)
```

### Learn the MSPN structure

```python
mspn = learn_mspn(train_data, ds_context, min_instances_slice=200, threshold=0.4)
```

### Show statistics of the learnt structure

```python
print(get_structure_stats(mspn))
```

### Plot the network structure

```python
plot_spn(mspn, 'lucas_spflow.png')
```
